Створити шлях до якого-небудь власного файлу з текстом програми. (Підставити власний шлях.)

```python
sample.stat().st_size
```

Знайдемо кількість рядків файлу

```python
if sample.stat().st_size < 2**20:
    lines = 0
    with open(sample) as f:
        for _ in f:  # ітеруємо рядками    _  тут ім'я змінної, що обходить рядки файлу, можна змінити
            lines += 1
    print(lines)
```

### Як зберегти табличну інформацію в csv-файл

Приклад даних. Табличні дані (як список контейнерів). (Використовувати списки чи кортежі тут різниці не має.) Та перелік заголовків до таблиці.

```python
data = [(2025, 1, 13, 23456, 4.5, 99, 'текст1'), 
        (2025, 7, 35, 23456, 4.5, 99, 'текст2'), 
        (2025, 4, 34, 23456, 4.5, 99, 'текст3'), 
        (2025, 11, 24, 23456, 4.5, 99, 'текст1')]
headers = ['h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'h7']
```

Роботу виконуватиме бібліотека <csv> (<a href="https://docs.python.org/3/library/csv.html" target=_blank 
rel="noopener noreferrer">документація</a>), більш конкретно <code>writer</code> (<a href="https://docs.python.org/3/library/csv.html#csv.writer" target=_blank 
rel="noopener noreferrer">документація</a>) . Для читання є <code>reader</code> (<a href="https://docs.python.org/3/library/csv.html#csv.reader" target=_blank 
rel="noopener noreferrer">документація</a>).

```python
import csv
```

```python
with open('data.csv', 'w', encoding='utf-8', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(headers)
    for row in data:
        writer.writerow(row)
```

```python
with open('data.csv', encoding='utf-8', newline='') as f:
    reader = csv.reader(f)
    headers = next(reader)
    lst = list()
    for row in reader:
        lst.append(row)
        print(row)
```
